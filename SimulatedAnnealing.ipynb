```python
from random import random
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ackley(x, a=20, b=0.2, c=2*math.pi):
    return -a*math.exp(-b*x) - math.exp(math.cos(c*x)) + a + math.e

x = list(np.arange(-50, 50, 0.1))

y = [ackley(i) for i in x]
plt.plot(x, y, color='green',label = 'Ackley function')
plt.legend()
plt.show()
```

```python
def generateNew(x,T):
    xnew = x + (random()-random()) * T
    return xnew
    
```

```python
def SA(f, T, N, alpha, x):
    Tlist = []
    ylist = []
    for i in range(1,N+1):
        Tlist.append(T)
        ylist.append(f(x))
        xnew = generateNew(x,T)
        deltay = f(xnew) - f(x)
        
        if deltay <= 0:
            x = xnew 
        else:
            metropolis = math.exp(-deltay/T)
            if random() < metropolis:
                x  = xnew
        T = T*alpha
    return Tlist,ylist
        
    
        
```

```python
Tlist1,ylist1 = SA(ackley, 20,1000,0.9,10)
Tlist2,ylist2 = SA(ackley, 20,1000,0.7,10)
Tlist3,ylist3 = SA(ackley, 20,1000,0.5,10)
```

```python
it = [i for i in range(1,1000+1)]
plt.plot(it,ylist1,label = 'alpha=0.9')
plt.plot(it,ylist2,label = 'alpha=0.7')
plt.plot(it,ylist3,label = 'alpha=0.5')
plt.xscale("log")
plt.xlabel('iteration')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
plt.plot(it,Tlist1,label = 'alpha=0.9')
plt.plot(it,Tlist2,label = 'alpha=0.7')
plt.plot(it,Tlist3,label = 'alpha=0.5')
plt.xscale("log")
plt.xlabel('iteration')
plt.ylabel('T')
plt.legend()
plt.show()
```
